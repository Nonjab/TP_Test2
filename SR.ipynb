```python
import pandas as pd
import seaborn as sns
import numpy as np

file_path = "C:/Users/DELL/Downloads/API_EN.POP.SLUM.UR.ZS_DS2_en_csv_v2_289192.csv"
data = pd.read_csv(file_path, skiprows=4)


print(data.info())
print(data.head())

relevant_columns = ['Country Name', 'Country Code', 'Indicator Name', 'Indicator Code', '2012']
data = data[relevant_columns]


missing_data = data.isnull().sum()
print("Missing data in each column:\n", missing_data)


data = data.dropna(subset=['2012'])

print(data.info())
print(data.describe())


data.to_csv("cleaned_slum_data.csv", index=False)
```
